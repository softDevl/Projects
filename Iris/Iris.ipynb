## Model

```python
df.drop('total_values',axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#label encode the target variable

encode=LabelEncoder()
df.species=encode.fit_transform(df.species)
df
```

```python
# train-test-split

train,test=train_test_split(df,test_size=0.2,random_state=0)
```

```python
print('shape of the training data:', train.shape)
print('shape of testing data:', test.shape)
```

```python
#seperate target and independent variable
train_x=train.drop(['species'],axis=1)
train_y=train['species']
```

```python
test_x = test.drop(columns=['species'],axis=1)
test_y = test['species']

outPut=pd.DataFrame([],columns=['Model','Accuracy'])
```

# Logistic Regression

```python
#create the object of the model

model=LogisticRegression()
model.fit(train_x,train_y)

predict=model.predict(test_x)
print('Predicted Values on Test Data',encode.inverse_transform(predict))

print('\n\nAccuracy Score on test data : \n\n')
acc_lr=accuracy_score(test_y,predict)
row = ['LR', acc_lr]
outPut.loc[len(outPut)] = row
outPut
```

# K-NN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(train_x,train_y)

predict=knn.predict(test_x)

accuracy_score(test_y,predict)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_GNB=GaussianNB()
model_GNB.fit(train_x,train_y)

predict=model_GNB.predict(test_x)

accuracy_score(test_y,predict)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_RF=RandomForestClassifier()
model_RF.fit(train_x,train_y)

predict=model_RF.predict(test_x)

accuracy_score(test_y,predict)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()

clf.fit(train_x,train_y)

predict=clf.predict(test_x)

accuracy_score(test_y,predict)
```

# SVM

```python
from sklearn.svm import LinearSVC
```

```python
lin_clf=LinearSVC()

lin_clf.fit(train_x,train_y)

predict=lin_clf.predict(test_x)

accuracy_score(test_y,predict)
```
